### Variable Arguements Through *args

```python
def add(*args):
    sum=0.0;
    for ele in args:
        sum+=ele;
    return sum
```

```python
print(add(3,4,6.0))
```

# Higher Order Functions Using Map(), Filter()....

## Maps

```python
li=[2,3,6,4,2,1]
```

```python
def addOne(number):
    return number+1

mapObject=map(addOne,li)
print(list(mapObject))
```

#### Lamda

```python
squared=list(map(lambda x:x**2,li))
print(squared)
```

## Filter

```python
def isEven(number):
    return number%2==0

filteredList=list(filter(isEven,li))
print(filteredList)


    
```
