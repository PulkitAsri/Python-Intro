```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris=sns.load_dataset("iris")
```

```python
iris
```

```python
sns.displot(iris["petal_length"],bins=20)
plt.show()
```

```python
sns.kdeplot(iris["petal_length"])
plt.show()
```

```python
sns.scatterplot(x="petal_length", y="petal_width", data=iris, hue="species")
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="hex")
plt.show()
```

```python
sns.pairplot(data=iris)
plt.show()
```

```python
sns.pairplot(data=iris, hue="species")
plt.show()
```

# Categorial plots

```python
sns.countplot(x="species", data=iris)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x="species", y="petal_length", data=iris)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(x="species", y="petal_length", data=iris)
plt.show()
```

```python
iris.corr()
```

```python
sns.heatmap(iris.corr(), cmap="Blues", annot=True)
```

## Visualise Tips Dataset

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.countplot(x="sex", data=tips)
plt.show()
```

```python
tips["day"].unique()
```

```python
sns.barplot(x="day",y="total_bill", data=tips, estimator=np.mean)
plt.show()
# mean by default, can be changes to custom fucntion, std dev, etc
```

```python
sns.barplot(x="day",y="total_bill", data=tips, hue="sex")
plt.show()
```

```python
sns.barplot(x="time",y="total_bill", data=tips, hue="day")
plt.show()
```

```python
sns.barplot(x="smoker",y="tip", data=tips)
plt.show()
```

```python
sns.scatterplot(x= "total_bill", y="tip", data=tips)
plt.show()
```
