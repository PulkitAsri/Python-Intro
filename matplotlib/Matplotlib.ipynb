## Bar Graph

```python
x=np.arange(4)
car1=[40,17,28,15]
car2=[30,37,18,11]
```

```python
plt.bar(x, car1, width=0.3, tick_label=[2017,2018,2019,2020],label="car1")
plt.bar(x+0.3, car2, width=0.3, label="car2")
plt.legend()
plt.xlabel("years")
plt.ylabel("Sales")
plt.xlim(-1,5)
plt.show()
```

## SubPlots

```python
x=np.arange(0,10,0.1)
x
y1=x**2
y2=x**3
y3=np.sin(x)
y4=np.cos(x)
```

```python
# plt.subplot(rows, cols, index)

plt.subplot(2, 2, 1)
plt.plot(x, y1)

plt.subplot(2, 2, 2)
plt.plot(x, y2)

plt.subplot(2, 2, 3)
plt.plot(x, y3)

plt.subplot(2, 2, 4)
plt.plot(x, y4)

plt.subplot(2, 2, 1)
plt.plot(x, y4)

plt.show()
```

## 3D Plots

```python
a=np.arange(4)
b=np.arange(4,7)

print(a)
print(b)
```

```python
a,b=np.meshgrid(a,b) # brodcast for making a mesh 

```

```python
print(a)
print(b)
print(a+b)
```

```python
fig=plt.figure()
ax=fig.gca(projection="3d")
ax.plot_surface(a,b,np.array([[6]]))
plt.show()
```

```python
a = np.arange(-1,1,0.005)
b = a
a, b=np.meshgrid(a,b)
```

```python
plt.figure().gca(projection="3d").plot_surface(a**2,b**2,(a**2 + b**2)**0.5)
plt.show()
```

## Reading From a CSV file

```python
import pandas as pd
```

```python
iris=pd.read_csv("./iris.csv")
```

```python
iris.head()
```

```python
iris["species"].unique()
```

```python
petal_length=iris['petal_length']
petal_width=iris['petal_width']
```

```python
plt.scatter(petal_length, petal_width)
plt.xlabel("ptal_length")
plt.ylabel("petal_width")
plt.show()
```

```python
setosa=iris[iris["species"]=="setosa"]
virginica=iris[iris["species"]=="virginica"]
versicolor=iris[iris["species"]=="versicolor"]
```

```python
plt.scatter(setose["petal_length"], setosa["petal_width"], color="green", label="setosa")
plt.scatter(virginica["petal_length"], virginica["petal_width"], color="red", label="virginica")
plt.scatter(versicolor["petal_length"], versicolor["petal_width"], color="blue",  label="versicolor")

plt.xlabel("ptal_length")
plt.ylabel("petal_width")
plt.legend()

plt.show()
```

```python
plt.hist(iris["petal_length"],bins=20)
plt.show()
```

```python
plt.pie(iris["species"].value_counts(), labels=iris["species"].value_counts().index)
plt.show()
```
